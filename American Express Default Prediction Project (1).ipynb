```python
# finidng variables' correlation
X1 = df1[['D_39', 'B_1', 'B_4', 'B_5', 'D_47', 'B_7', 'D_51', 'B_9', 'B_10', 'B_11', 'B_12', 'D_58', 'B_14', 'D_60', 'D_63', 'D_65', 'B_18', 'B_21', 'D_71', 'B_23', 'D_75', 'B_24', 'B_28', 'D_86', 'B_31', 'B_32', 'D_92', 'D_93', 'D_94', 'D_96', 'B_36', 'D_127']]
corr_matrix = X1.corr()
corr_matrix.head()
```

```python
upper_tri = corr_matrix.where(np.triu(np.ones(corr_matrix.shape),k=1).astype(np.bool))
print(upper_tri)
```

```python
# finding variables with a correlation greater than 72%
highest = [column for column in upper_tri.columns if any(upper_tri[column] > 0.72)]
print(highest)
```

```python
# selecting 2000 rows to use the model on 
df1.sample(n = 2000)
```

```python
# X variables and y variable (target)
X = df1[['B_11', 'D_58', 'B_14', 'B_23', 'D_75', 'B_28']]
y = df1.target
X.shape

# split training set and testing set 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.25)

# Running Logistic Regression model
model = sklearn.linear_model.LogisticRegression()
model.fit(X_train,y_train)
y_pred = model.predict(X_test)
print(y_pred)
```

```python
# Accuracy of Model
from sklearn.metrics import classification_report
matrix = classification_report(y_test, y_pred)
print(matrix)
```

# Our Process

### Why did we decided to only use (D_* ) Delinquency variables & (B_* ) Balance variables?

For the spending variables, we thought those would be unneccessary because we have the balance variables that tell us how much the customers owe after spending. We got rid of the payment variables for the same reason because balance variables will tell us the end result of what a customer needs to pay still. Risk variables could have been helpful for a prediction but delinquency variables seemed like the stronger choice for the prediction model because it tells us a history of delinquency and what could lead to a customer not filling a payment. For example, 'D_58' could have been the percentage of payments that weren't paid on time within the last 5 years. With 190 different variables and over 5.5 million customers, we needed to slim down most of the variables in order to process the program.

### How else did we clean the data?

We started by weeding out the unwanted variables (S, P, and R). Then, deleted columns with any null values because we wanted columns that had the most information to give for the prediction. Afterwards, we merged the train_labels csv file to the training dataset because we needed the 'target' variable (y) that described whether a default event would or would not occur for the predictive model. Once finding the variables with the highest correlation, we ran a logistic regression to see if a customer paid or didn't pay in which target = 1 meant they don't pay within the 120 days and target = 0 meant they do pay. 
Note: We didn't have to encode any of the categorical variables because the correlation for the variables didn't label any categorical variables as variables with the highest correlation.
